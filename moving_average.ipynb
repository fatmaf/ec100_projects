```python
#importing a library to plot stuff for visualization only 
import matplotlib.pyplot as plt
```

```python
data=[3,2,5,2,6,65,23,22,1,2,5,6]
data_x = range(len(data))
window_size = 3
```

```python
#the function that does the moving average 
#data = the data 
#window_size = number of samples to average over 
#
def do_moving_average(data,window_size):
    cumulative_sum = [0] #save the cumulative sum of every window_size 
    # go over all the data points 
    new_data = []
    for i in range(len(data)):
        
        cumulative_sum.append(cumulative_sum[i]+data[i])
        if i >= (window_size-1):
            averaged_data = (cumulative_sum[i+1]-cumulative_sum[i])/window_size
            new_data.append(averaged_data)
    #print(data)
    #print(cumulative_sum)    
    #print(new_data)
    return new_data
            
```

```python
plt.plot(data_x,data)
```

```python
new_data = do_moving_average(data,window_size)
#print(new_data)
```

```python
plt.plot(data_x,data)
plt.plot(range(window_size-1,len(data)),new_data)
```
